```python
DATAPATH='data/'  # must end in "/"
sys.path.append("..") # append parent dir in order to use sibling dirs

from SimTools.ORF_eliminator import *
from SimTools.RNA_gen import *
from SimTools.RNA_describe import *
```

```python
NO_OF_SEQUENCES = 5
BASES = ['A', 'C', 'T', 'G']
```

```python
describer = ORF_RE()
```

```python
generator = Collection_Generator()
list_of_random_sequence = generator.get_sequences(NO_OF_SEQUENCES)
list_of_bases = []

for list in list_of_random_sequence:
    number_bases = describer.get_number_bases(list)
    list_of_bases.append(number_bases)
    
```

```python
eliminator = ORF_eliminator()
list_of_bases_noORF=[]
for list in list_of_random_sequence:
    rna = eliminator.eliminate_ORF(list)
    number_bases_noORF = describer.get_number_bases(rna)
    list_of_bases_noORF.append(number_bases_noORF)
    
```

```python
counter =0
while (counter<NO_OF_SEQUENCES):
    data1 = list_of_bases[counter]
    data2 = list_of_bases_noORF[counter]
    width = 0.3
    plt.bar(np.arange(len(data1)), data1, width=width)
    plt.bar(np.arange(len(data2))+ width, data2, width=width)
    plt.xlabel(BASES)
    
    
    plt.show()
    counter+=1
```
